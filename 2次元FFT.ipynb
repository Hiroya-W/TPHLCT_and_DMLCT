```python
import numpy as np
import scipy.misc
from scipy.fftpack import dct, idct
import sys
from PIL import Image
import matplotlib.pyplot as plt
```

```python
bmp = Image.open("LENNA.bmp")
plt.imshow(bmp,"gray")
```

```python
# 2FFT
lenna_F = np.fft.fft2(bmp)
```

```python
# 
plt.imshow(np.log10(np.abs(lenna_F)))
```

```python
# 
# 1423
lenna_F = np.fft.fftshift(lenna_F)
plt.imshow(np.log10(np.abs(lenna_F)))
```
